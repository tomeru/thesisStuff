```python
from IPython.html.widgets import interact

ypos = 3.5
circleSize = 1.5

position1 = [8, 10, 12, 12, 12, 12]
position2 = [15, 15, 15, 17, 19, 21]

groundPoints = [[0, 0], [0, 2], [xLim, 2], [xLim, 0]]
occluderPoints = [[0, 2], [0, 8], [4, 8], [4, 2]]
xLim = 30
```

```python
def runAnim(frame=0):
    fig = figure()
    #fig.set_dpi(10)
    fig.set_size_inches(15, 5)
    ax = axes(xlim=(0, xLim), ylim=(0, 10))

    patch1 = Circle((position1[frame], ypos), circleSize, fc='b')
    patch2 = Circle((position2[frame], ypos), circleSize, fc='r')

    groundLine = plt.Polygon(groundPoints, edgecolor='k', facecolor='k')
    occluderLine = plt.Polygon(occluderPoints, edgecolor='k', facecolor='gray')

    ax.add_patch(patch1)
    ax.add_patch(patch2)
    ax.add_line(groundLine)
    ax.add_line(occluderLine)
    
    
interact(runAnim, frame=(0,5,1));
#runAnim(2)
```
